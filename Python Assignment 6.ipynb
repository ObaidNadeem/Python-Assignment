```python
#Question 1: Define Object Oriented Programming Language?
#Answer 1: Programming language in which object can be defined as data field. In OOP real world objects can be defined.

#Question 2: List down the Benefits of OOP?
#Answer 2: 1. You can find bugs easily.
#          2. Code is reuseable.
#          3. Others can understand your code.

# Question 3: Differentiate between function and method?
# Answer 3: When a piece of code is called by name it is a "function"
#           And if that piece of code is associated with an object then it is called "Method"

#Question 4: Define the following terms:
#           1. Class: A class is a user defined blueprint or prototype from which objects are created. In which you can save the objects.
#           2. Object: An object is collection of data and method in Python.
#           3. Attribute: An attribute is a specification that defines a property of an object.
#           4. Behavior: It is one of the most popuplar BDD framework in Python.

'''
Question 5: Write a code in python in which create a class named it Car which
have 5 attributes such like (model, color and name etc.) and 3
methods. And create 5 object instance from that class.

Answer 5:
'''




class Car():
    
    def __init__(self, model, colour, name, marketprice, mileage):
        self.model = model
        self.colour = colour
        self.name = name
        self.marketprice = marketprice
        self.mileage = mileage
        
    
    #Updates
    
    def update_name(self, newname):
        self.name = newname
    
    
    def update_color(self, newcolor):
        self.color = newcolor


        
# First instance

car1 = Car("2018", "Black", "Corolla", "30 lacs", "475 miles")
print ("Car 1 attributes:")
print ("- - - - - - - - - ")
print("This car was made in",car1.model)
print("The colour of this car is",car1.colour)
print("The name of this car is",car1.name)
print("The market price of this car is",car1.marketprice)
print("This car can go",car1.mileage, "on refueling full tank")
print ("- - - - - - - - - " + "\n")

# Second instance

car2 = Car("2017", "Black", "Vigo", "50 lacs", "900 miles")
print ("Car 2 attributes:")
print ("- - - - - - - - - ")
print("This car was made in",car2.model)
print("The colour of this car is",car2.colour)
print("The name of this car is",car2.name)
print("The market price of this car is",car2.marketprice)
print("This car can go",car2.mileage, "on refueling full tank")
print ("- - - - - - - - - " + "\n")

# Third instance

car3 = Car("2019", "White" , "Civic", "35 lacs", "400 miles")
print ("Car 3 attributes:")
print ("- - - - - - - - - ")
print("This car was made in",car3.model)
print("The colour of this car is",car3.colour)
print("The name of this car is",car3.name)
print("The market price of this car is",car3.marketprice)
print("This car can go",car3.mileage, "on refueling full tank")
print ("- - - - - - - - - " + "\n")

# Fourth instance

car4 = Car("2012", "Brown" , "Mehran", "7 lacs", "150 miles")
print ("Car 4 attributes:")
print ("- - - - - - - - - ")
print("This car was made in",car4.model)
print("The colour of this car is",car4.colour)
print("The name of this car is",car4.name)
print("The market price of this car is",car4.marketprice)
print("This car can go",car4.mileage, "on refueling full tank")
print ("- - - - - - - - - " + "\n")

# Fifth instance

car5 = Car("2018", "White" , "Coure", "10 lacs", "100 miles")
print ("Car 5 attributes:")
print ("- - - - - - - - - ")
print("This car was made in",car5.model)
print("The colour of this car is",car5.colour)
print("The name of this car is",car5.name)
print("The market price of this car is",car5.marketprice)
print("This car can go",car5.mileage, "on refueling full tank")
print ("- - - - - - - - - " + "\n")



    
print("Car previously owned was",car5.name)
car5.update_name("Vigo")
print("New car brought is",car5.name)
```
